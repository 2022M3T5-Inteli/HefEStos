#Importação das bibliotecas e arquivos

```python
#Esta parte descreve a importação de bibliotecas e de arquivos do drive
import matplotlib.pyplot as plt
import pandas as pd
import datetime as dt
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
from matplotlib.pylab import rcParams

#Importação do arquivo do drive
from google.colab import drive
drive.mount('/content/drive')
dataSet = pd.read_csv("/content/drive/Shareddrives/hefestos/CSV/semilimpa_nao_arredondado.csv", sep = ',')

#Ajustes no dataset da emissora 1 de segunda a sexta e sábado e domingo
dataSet = dataSet.drop(['Programa'], axis=1)
test = pd.get_dummies(dataSet['Categoria'], drop_first=False)
Categorized = pd.concat([dataSet, test], axis = 1)
Categorized = Categorized.drop(['Categoria','Unnamed: 0','Unnamed: 0_x',], axis=1)
Categorized
```

```python
#Criação de um novo dataset a partir das modificações feitas
Categorized.to_csv('/content/drive/Shareddrives/hefestos/CSV/Tabela_limpa_completa_float.csv')
```

#Escolha das Features

O valor da correlação varia entre -1 e 1, e quanto mais próximo desses extremos mais forte será a correlação entre as variáveis, seja ela positiva ou negativa.

```python
# Definindo tamanho da figura
rcParams['figure.figsize'] = 60, 40
# Matriz de correlação
matriz_de_correlacao = Categorized.corr()
# Mapa de calor
sns.heatmap(matriz_de_correlacao, annot=True, vmin=-1, vmax=1, center=0)
# Definindo a posição dos ticks nos eixos
plt.yticks(rotation=90)
plt.xticks(rotation=180)
# Mostrando a figura
plt.show()  
```
